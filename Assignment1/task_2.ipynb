# Problem 2

Here, we will use python to find the roots of the function => cos(x^(1/2)) + x^2 - 7x.

For all of the methods, iterations shall continue until the approximate percent relative error is less than 0.001% 

```python
import numpy as np
```

```python
# Define the function f(x) = cos(sqrt(x)) + x^2 - 7x
def f(x):
    return np.cos(np.sqrt(x)) + np.power(x, 2) - 7 * x
```

```python
# Define the derivative f'(x) = -(1/(2*sqrt(x)))*sin(sqrt(x)) + 2x - 7
def df(x):
    return -(1 / (2 * np.sqrt(x))) * np.sin(np.sqrt(x)) + 2 * x - 7
```

```python
# Function to calculate approximate percent relative error
def approximate_percent_relative_error(x_new, x_old):
    return abs((x_new - x_old) / x_new) * 100
```

## (a) Bisection Method



```python
# The initial guesses
xl = 6.0  # lower guess
xu = 8.0  # upper guess

xr = (xl + xu) / 2.0  # initial root guess

itearations = 1

while True:
    xr_old = xr
    fxl = f(xl)
    fxu = f(xu)
    fxr = f(xr)

    if fxl * fxr < 0:
        xu = xr
    elif fxu * fxr < 0:
        xl = xr
    else:
        break

    xr = (xl + xu) / 2.0

    ea = approximate_percent_relative_error(xr, xr_old)

    print(f"Iteration: {itearations}, Approximate Percent Relative Error: {ea}%")

    # If relative error is less than 0.001%, stop the iteration
    if ea < 0.001:
        break

    itearations += 1

print(f"Root found: {xr} after {itearations} iterations")
```

## (b) Newton-Raphson Method 

```python
# The initial guesses
x0 = 6.0

itearations = 1

while True:
    x0_old = x0
    f_x0 = f(x0)
    df_x0 = df(x0)

    x0 = x0 - f_x0 / df_x0

    ea = approximate_percent_relative_error(x0, x0_old)

    print(f"Iteration: {itearations}, Approximate Percent Relative Error: {ea}%")

    # If relative error is less than 0.001%, stop the iteration
    if ea < 0.001:
        break

    itearations += 1

print(f"Root found: {x0} after {itearations} iterations")
```

## Secant Method

```python
# The initial guesses
x0 = 6.0
x1 = 5.0

itearations = 1

while True:
    x0_old = x0
    f_x0 = f(x0)
    f_x1 = f(x1)

    x_temp = x0 - f_x0 * (x1 - x0) / (f_x1 - f_x0)
    x1 = x0
    x0 = x_temp

    ea = approximate_percent_relative_error(x0, x0_old)

    print(f"Iteration: {itearations}, Approximate Percent Relative Error: {ea}%")

    # If relative error is less than 0.001%, stop the iteration
    if ea < 0.001:
        break

    itearations += 1

print(f"Root found: {x0} after {itearations} iterations")
```
