# Problem 2

Here we will solve a set of equations using the Gauss-Seidel method and the Jacobi Method.

The equations are as follows:

x + 2y - 4z + 2t = 108.6 ....(1)

-x + 3y - 2z + u = 92.3 ....(2)

-2y + 3z - 6t = 98.5 ....(3)

-3x - z + 2t - 7u = 102.7 ....(4)

3x + y - t + 2u = 111.2 ....(5)



```python
import numpy as np
```

```python
# These are the equations that we are trying to solve for x, y, z, t, and u.
# x + 2y - 4z + 2t = 108.6 ....(1)

# -x + 3y - 2z + u = 92.3 ....(2)

# -2y + 3z - 6t = 98.5 ....(3)

# -3x - z + 2t - 7u = 102.7 ....(4)

# 3x + y - t + 2u = 111.2 ....(5)


# Note that in order for the Gauss-Seidel method to converge, the coefficient matrix of the system of equations should be diagonally dominant. In this case, we can rearrange the equations to make them diagonally dominant.

def value_of_x_from_equation_5(y, t, u):
    return (-y + t - 2 * u + 111.2) / 3


# Value of y from equation (2): -x + 3y - 2z + u = 92.3
def value_of_y_from_equation_2(x, z, u):
    return (x + 2 * z - u + 92.3) / 3


# Value of z from equation (1): x + 2y - 4z + 2t = 108.6
def value_of_z_from_equation_1(x, y, t):
    return (x + 2 * y + 2 * t - 108.6) / 4


# Value of t from equation (3): -2y + 3z - 6t = 98.5
def value_of_t_from_equation_3(y, z):
    return (-2 * y + 3 * z - 98.5) / 6


# Value of u from equation (4): -3x - z + 2t - 7u = 102.7
def value_of_u_from_equation_4(x, z, t):
    return (-3 * x - z + 2 * t - 102.7) / 7
```

```python
# Function to calculate approximate percent relative error
def approximate_percent_relative_error(x_new, x_old):
    return abs((x_new - x_old) / x_new) * 100
```

## Gauss-Seidel Method

```python
# Initial guess for x, y, z, t, and u.
x, y, z, t, u = 0, 0, 0, 0, 0

# Relative percent error threshold for convergence.
relative_error_threshold = 0.0002

# Initial relative errors for x, y, z, t, and u.
error_x, error_y, error_z, error_t, error_u = 100, 100, 100, 100, 100

iterations = 0

# Iteratively solve for x, y, z, t, and u until convergence. It should iterate until all relative errors are less than the threshold.
while (
    error_x > relative_error_threshold
    or error_y > relative_error_threshold
    or error_z > relative_error_threshold
    or error_t > relative_error_threshold
    or error_u > relative_error_threshold
):
    iterations += 1
    # Store the old values of x, y, z, t, and u for error calculation.
    old_x, old_y, old_z, old_t, old_u = x, y, z, t, u

    # Update x, y, z, t, and u using the equations.
    x = value_of_x_from_equation_5(y, t, u)
    y = value_of_y_from_equation_2(x, z, u)
    z = value_of_z_from_equation_1(x, y, t)
    t = value_of_t_from_equation_3(y, z)
    u = value_of_u_from_equation_4(x, z, t)

    # Calculate the relative errors for x, y, z, t, and u.
    error_x = approximate_percent_relative_error(x, old_x)
    error_y = approximate_percent_relative_error(y, old_y)
    error_z = approximate_percent_relative_error(z, old_z)
    error_t = approximate_percent_relative_error(t, old_t)
    error_u = approximate_percent_relative_error(u, old_u)
    # Print the current values of x, y, z, t, u and their relative errors.
    print(f"Iteration {iterations}:")
    print(
        f"Relative Errors - x: {error_x:.4f}%, y: {error_y:.4f}%, z: {error_z:.4f}%, t: {error_t:.4f}%, u: {error_u:.4f}%\n"
    )
# Print the final values of x, y, z, t, and u after convergence.
print()
print(f"Converged in {iterations} iterations.")
print(
    f"Final values after convergence - x: {x:.4f}, y: {y:.4f}, z: {z:.4f}, t: {t:.4f}, u: {u:.4f}"
)
```

```python
# Verify the solution by plugging the final values back into the original equations.
equation_1 = x + 2 * y - 4 * z + 2
equation_2 = -x + 3 * y - 2 * z + u
equation_3 = -2 * y + 3 * z - 6 * t
equation_4 = -3 * x - z + 2 * t - 7 * u
equation_5 = 3 * x + y - t + 2 * u

print("\nVerification of the solution:")
print(f"Equation 1: {equation_1:.4f} (should be close to 108.6)")
print(f"Equation 2: {equation_2:.4f} (should be close to 92.3)")
print(f"Equation 3: {equation_3:.4f} (should be close to 98.5)")
print(f"Equation 4: {equation_4:.4f} (should be close to 102.7)")
print(f"Equation 5: {equation_5:.4f} (should be close to 111.2)")
```
